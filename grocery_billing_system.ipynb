```python
grocery_inventory = ["apple", "banana", "orange", "milk", "cheese", "eggs", "sugar", "cabbage", "cereal", "rice", "mushrooms" ] #more items can be added to the inventory
prices = {"apple": 50, "banana" : 80, "orange": 30, "milk" : 40, "cheese" : 95, "eggs": 120, "sugar" : 25, "cabbage" : 10, "cereal" : 30, "rice" : 25, "mushrooms" : 45 }
retail_inq = int(input("enter amount of items you wanna buy"))
price= 0
total_items= 0
invalid_items= 0
no_of_items = 0

def inventorycheck(no_of_items):
    i=0
    global price, total_items, invalid_items
    while i< no_of_items:
        item_name = input("input item name")
        i= i+1
        if item_name in grocery_inventory:
            print("item added successfully")
            price = price + prices[item_name]
            total_items = total_items +1
        else:
            print("sorry item not available")
            invalid_items = invalid_items +1
inventorycheck(retail_inq)
if invalid_items > 0:
    recheck_items = input("do u wish to add something else?")
    if recheck_items == "yes":
        inventorycheck(invalid_items)
        invalid_items = 0
        print("your total bill is", price, "total no. of items:", total_items)
    else:
        print("your total bill is", price, "total no. of items:", total_items)
member = input("do you have our gold membership?")
if member == "yes":
    customer_id = int(input("enter ur relationship id"))
    if customer_id == 1234:
        price = price - 20/100 * price
        print("your new bill is", price)
    else:
        print("invalid user id")
        recheck_id = input("do u want to try again?")
        if recheck_id == "yes":
            customer_id2 = int(input("enter relationship id"))
            if customer_id2 == 1234:
                price = price * 0.8
                print("success,,", price, "= ur total bill")
else:
    print(price, "checkout now >")
print("thank you for shopping with us!")

```
